# अभ्यास ५ - परिवर्तनशील क्वांटम ईगेनसॉल्वर


## ऐतिहासिक परिप्रेक्ष्य 


पिछले दशक के दौरान, क्वांटम कंप्यूटर तेजी से उन्नति हासिल की और कंप्यूटिंग प्रणाली जो क्वांटम तरीके से प्रकृति के नियमों का अनुकरण कर सकती है जो फेनमैन का प्रारंभिक सपना था वह साकार होने लगा। अल्बर्टो पेरुज़ो द्वारा पहली बार लिखे गए २०१४ के एक पेपर ने **परिवर्तनशील यानी वैरिएशनल क्वांटम ईगेनसॉल्वर (वीक्यूई/VQE)** पेश किया, एक एल्गोरिथ्म जो एक अणु की जमीनी अवस्था ऊर्जा (सबसे कम ऊर्जा) को खोजने के लिए इस्तेमाल किया जाता है, जिसमें अन्य दृष्टिकोणों की तुलना में बहुत उथले सर्किट होते हैं|[१] और २०१७ में, क्वांटम टीम ने लिथियम हाइड्राइड अणु की जमीनी अवस्था ऊर्जा का अनुकरण करने के लिए 
आईबीएम (IBM) ने वीक्यूई एल्गोरिथ्म का उपयोग किया|[२]

वीक्यूई का जादू समस्या के कुछ प्रसंस्करण कार्यभार को शास्त्रीय कंप्यूटर पर सौंपने से आता है| एल्गोरिथ्म एक पैरामीटरयुक्त क्वांटम सर्किट से शुरू होता है जिसे अंसात्ज़(एक सर्वोत्तम अनुमान) कहा जाता है, फिर शास्त्रीय अनुकूलक का उपयोग करके इस सर्किट के लिए इष्टतम पैरामीटर ढूंढता है। शास्त्रीय एल्गोरिथ्म पर वीक्यूई का लाभ इस तथ्य से आता है कि एक क्वांटम प्रसंस्करण इकाई समस्या की सटीक तरंग का प्रतिनिधित्व और भंडारण कर सकती है, वही जो एक शास्त्रीय कंप्यूटर के लिए घातीय रूप से कठिन समस्या हैं। 

यह अभ्यास ५ आपको फेनमैन के सपने को स्वयं साकार करने की अवसर देता है, जमीन की स्थिति और एक अणु की ऊर्जा को निर्धारित करने के लिए एक परिवर्तनशील क्वांटम ईजेन्सोलवर की स्थापना करता है। यह दिलचस्प है क्योंकि जमीनी अवस्था का उपयोग विभिन्न आणविक गुणों की गणना के लिए किया जा सकता है, उदाहरण के लिए नाभिक पर सटीक बल आणविक गतिशीलता सिमुलेशन चलाने के लिए सेवा कर सकते हैं ताकि यह पता लगाया जा सके कि समय के साथ रासायनिक प्रणालियों में क्या होता है।

## परिचय

वीक्यूई के कार्यान्वयन के लिए, विशेष रूप से अंसात्ज़ क्वांटम सर्किट पर ध्यान केंद्रित करते हुए, आप अपने सिमुलेशन की रचना किस तरह करना चाहते हैं इस पर चुनाव करने में सक्षम होंगे| यह इस तथ्य से प्रेरित है कि शोर क्वांटम कंप्यूटरों पर वीक्यूई चलाते समय महत्वपूर्ण कार्यों में से एक जमीनी स्थिति का प्रतिनिधित्व करने में सक्षम सबसे कॉम्पैक्ट क्वांटम सर्किट ढूंढकर विश्वसनीयता के नुकसान को सीमित करना है (जो त्रुटियों का परिचय देता है।

## अभ्यास ५: परिवर्तनशील क्वांटम ईगेनसॉल्वर 

## सिद्धांत

यहाँ नीचे सामान्य वर्कफ़्लो है जो दर्शाता है कि वीक्यूई का उपयोग करके आणविक सिमुलेशन क्वांटम कंप्यूटर पर कैसे किए जाते हैं|

<img src="resources/workflow.png" width=800 height= 1400/>

मूल विचार हाइब्रिड क्वांटम-शास्त्रीय दृष्टिकोण **सीपीयू (क्लासिकल प्रोसेसिंग यूनिट)** और **क्यूपीयू (क्वांटम प्रोसेसिंग यूनिट)** को उन हिस्सों को सौंपना है जो वे सबसे अच्छा कर सकते हैं| सीपीयू उन शर्तों को सूचीबद्ध करने का ध्यान रखता है जिन्हें ऊर्जा की गणना करने और सर्किट मापदंडों को अनुकूलित करने के लिए नापने की आवश्यकता होती है|

क्यूपीयू एक क्वांटम सर्किट को लागू करता है जो एक सिस्टम की क्वांटम स्थिति का प्रतिनिधित्व करता है और ऊर्जा को मनापता है। 
कुछ और विवरण नीचे दिए गए हैं:

**सीपीयू** इलेक्ट्रॉन हॉपिंग और अंतःक्रियाओं से जुड़ी ऊर्जाओं की कुशलता से गणना कर सकते हैं (हार्ट्री-फॉक गणना के माध्यम से एक-/दो-बोडी इंटीग्रल) जो कुल ऊर्जा ऑपरेटर, हैमिल्टनियन का प्रतिनिधित्व करने के लिए काम करते हैं| [हार्ट्री-फॉक(HF) गणना](https://en.wikipedia.org/wiki/Hartree%E2%80%93Fock_method#:~:text=In%20computational%20physics%20and%20chemistry,system%20in%20a%20stationary%20state.) एक अनुमानित ग्राउंड स्टेट वेवफंक्शन की कुशलता से गणना करता है यह मानते हुए कि तरंग फ़ंक्शन को स्लेटर निर्धारक द्वारा दर्शाया जा सकता है(उदाहरण के लिए, STO-3G आधार में H$_2$ अणु के लिए 4 स्पिन-ऑर्बिटल्स और क्यूबिट्स के संदर्भ में, $|\Psi_{HF} \rangle = |0101 \rangle$ जहां इलेक्ट्रॉन सबसे कम ऊर्जा स्पिन-ऑर्बिटल्स पर कब्जा करते हैं)| वीक्यूई के अगले चरण  रूप में, क्यूपीयू एक क्वांटम स्थिति (संबंधित सर्किट और इसके पैरामीटर) खोजता है जो अन्य स्थितियों से जुड़े लापता इलेक्ट्रॉनिक सहसंबंधों का भी प्रतिनिधित्व कर सकता है (अर्थात्, $\sum_i c_i |i\rangle$ स्थितियों में $|\Psi \rangle = c_{HF}|\Psi_{HF} \rangle + \sum_i c_i |i\rangle $ जहाँ $i$ 
एक बिटस्ट्रिंग है)|

एचएफ(HF) गणना के बाद, हैमिल्टनियन में ऑपरेटरों को क्यूपीयू पर फर्मियन-टू-क्यूबिट् ट्रांसफॉर्मेशन का उपयोग करके मापने के लिए मैप किया जाता है(नीचे हैमिल्टनियन अनुभाग देखें).  सिस्टम के गुणों का विश्लेषण हम कर सकते हैं ताकि क्यूबिट् की संख्या कम हो या अंसात्ज़ सर्किट छोटा हो:

- Z2 समरूपता और दो-क्यूबिट् कमी के लिए, देखें [Bravyi *et al*, 2017](https://arxiv.org/abs/1701.08213v1)|
- एंटांगलेमेंट फोर्जिंग के लिए, देखें [Eddins *et al.*, 2021](https://arxiv.org/abs/2104.10220v1)|
- अनुकूली अंसात्ज़ के लिए देखें, [Grimsley *et al.*,2018](https://arxiv.org/abs/1812.11173v2), [Rattew *et al.*,2019](https://arxiv.org/abs/1910.09694), [Tang *et al.*,2019](https://arxiv.org/abs/1911.10205). आप इन कार्यों में पाए गए विचारों का उपयोग क्वांटम सर्किट को छोटा करने के तरीके खोजने के लिए कर सकते हैं।


**क्यूपीयू(QPU)** क्वांटम सर्किट लागू करता है (अंसात्ज़ अनुभाग नीचे देखें), कोणों द्वारा पैरामीटरकृत $\vec\theta$, जो विभिन्न सिंगल क्यूबिट् घुमावों और उलझावों को रखकर ग्राउंड स्टेट वेवफंक्शन का प्रतिनिधित्व करेगा (जैसे की दो-क्यूबिट् गेटस). क्वांटम लाभ इस तथ्य में निहित है कि क्यूपीयू सटीक तरंग का कुशलतापूर्वक प्रतिनिधित्व और भंडारण कर सकता है, जो कुछ से अधिक परमाणुओं वाले सिस्टम के लिए शास्त्रीय कंप्यूटर पर दुथसाध्य हो जाता है|  अंत में, क्यूपीयू पसंद के ऑपरेटरों को मापता है (जैसे एक हैमिल्टन का प्रतिनिधित्व करने वाले)|

नीचे हम वीक्यूई एल्गोरिथम के प्रत्येक घटक के गणितीय विवरण में थोड़ा और आगे जाते हैं। अगर आप वीक्यूई के बारे में हमारा वीडियो देखते हैं तो यह भी मददगार हो सकता है [video episode about VQE](https://www.youtube.com/watch?v=Z-A6G0WVI9w)|


### हैमिल्टनियन


यहां हम किसी दिए गए सिस्टम की ऊर्जा को मापने के लिए आवश्यक ऑपरेटरों को प्राप्त करने की विधि की व्याख्या करते हैं।
इन शर्तों को आणविक हैमिल्टन में परिभाषित किया गया है:
$$
\begin{aligned}
\hat{H} &=\sum_{r s} h_{r s} \hat{a}_{r}^{\dagger} \hat{a}_{s} \\
&+\frac{1}{2} \sum_{p q r s} g_{p q r s} \hat{a}_{p}^{\dagger} \hat{a}_{q}^{\dagger} \hat{a}_{r} \hat{a}_{s}+E_{N N}
\end{aligned}
$$
with
$$
h_{p q}=\int \phi_{p}^{*}(r)\left(-\frac{1}{2} \nabla^{2}-\sum_{I} \frac{Z_{I}}{R_{I}-r}\right) \phi_{q}(r)
$$
$$
g_{p q r s}=\int \frac{\phi_{p}^{*}\left(r_{1}\right) \phi_{q}^{*}\left(r_{2}\right) \phi_{r}\left(r_{2}\right) \phi_{s}\left(r_{1}\right)}{\left|r_{1}-r_{2}\right|} 
$$

जहां $h_{r s}$  और $g_{p q r s}$ एक-/दो-बोडी इंटीग्रल है (हार्ट्री-फॉक विधि का उपयोग करके) और $E_{N N}$ परमाणु प्रतिकर्षण ऊर्जा हैं|
यह एक-बोडी इंटीग्रल इलेक्ट्रॉनों की गतिज ऊर्जा और नाभिक के साथ उनकी परस्पर क्रिया का प्रतिनिधित्व करते हैं| 
यह दो-बोडी इंटीग्रल इलेक्ट्रॉन-इलेक्ट्रॉन परस्पर क्रिया का प्रतिनिधित्व करते हैं|
यह $\hat{a}_{r}^{\dagger}, \hat{a}_{r}$ ऑपरेटर स्पिन-ऑर्बिटल में इलेक्ट्रॉन के निर्माण और विनाश का प्रतिनिधित्व करते हैं|
 $r$ और ऑपरेटरों को मैपिंग की आवश्यकता होती है, ताकि हम उन्हें क्वांटम कंप्यूटर पर माप सकें|
ध्यान दें कि वीक्यूई इलेक्ट्रॉनिक ऊर्जा को कम करता है इसलिए आपको कुल ऊर्जा की गणना करने के लिए परमाणु प्रतिकर्षण ऊर्जा $E_{NN}$ को पुनः प्राप्त करना और जोड़ना होगा।
    

तो, $h_{r s}$ और $g_{p q r s}$ टेंसर में प्रत्येक गैर-शून्य मैट्रिक्स तत्व के लिए, हम निम्नलिखित फर्मियन-से-क्यूबिट परिवर्तन के साथ संबंधित पाउली स्ट्रिंग (पॉली ऑपरेटरों के टेंसर उत्पाद) का निर्माण कर सकते हैं।
उदाहरण के लिए, जॉर्डन-विग्नर मैपिंग में एक कक्षीय $r = 3$ के लिए, हम निम्नलिखित पाउली स्ट्रिंग प्राप्त करते हैं:
$$
\hat a_{3}^{\dagger}= \hat \sigma_z \otimes \hat \sigma_z \otimes\left(\frac{ \hat \sigma_x-i \hat \sigma_y}{2}\right) \otimes 1 \otimes \cdots \otimes 1
$$
जहां $\hat \sigma_x, \hat \sigma_y, \hat \sigma_z$ प्रसिद्ध पाउली ऑपरेटर हैं।
$\hat \sigma_z$ ऑपरेटरों के टेंसर उत्पादों को फर्मोनिक एंटी-कम्यूटेशन संबंधों को लागू करने के लिए रखा गया है।

### अंसत्ज़ेस (Ansatzes)

रासायनिक समस्याओं के लिए आप मुख्य रूप से २ प्रकार के अंसत्ज़ेस का उपयोग कर सकते हैं।

- **क्यू-यूसीसी अंसत्ज़ेस** भौतिक रूप से प्रेरित हैं, और अपरिष्कृत तौर पर क्वांटम सर्किट में इलेक्ट्रॉन उत्तेजनाओं को मैप करते हैं| The क्यू-यूसीसीएसडी
अंसात्ज़ (किस्किट में`यूसीसीएसडी`) सभी संभव सिंगल और डबल इलेक्ट्रॉन उत्तेजना को धारणा  करते हैं | युग्मित डबल (`पीयूसीसीडी`) तथा सिंग्लेट क्यू-यूसीसीएसडी0 (`एसयूसीसीडी`) बस ऐसे उत्तेजनाओं के सबसेट पर विचार करते हैं (मतलब काफी कम सर्किट)और वियोजन प्रोफाइल के लिए अच्छे परिणाम प्रदान करने के लिए सिद्ध हुए हैं. उदाहरण के लिए, क्यू-पीयूसीसीडी में एकल उत्तेजना नहीं होती है और दोहरे उत्तेजनाओं को नीचे की छवि के अनुसार जोड़ा जाता है|
- **अनुमानी  अंसत्ज़ेस (`टूलोकल`/`TwoLocal`)** सर्किट की गहराई को छोटा करने के लिए आविष्कार किया गया था लेकिन फिर भी जमीन की स्थिति का प्रतिनिधित्व करने में सक्षम हैं| 

जैसा कि नीचे दिए गए चित्र में है, आर गेट्स पैरामीट्रिज्ड सिंगल क्यूबिट रोटेशन का प्रतिनिधित्व करते हैं और $U_{CNOT}$ एंटेंगलर्स को (दो-क्यूबिट गेट्स)| विचार यह है कि एक ही ब्लॉक (स्वतंत्र मापदंडों के साथ) को कुछ $D$-बार दोहराने के बाद कोई जमीनी स्थिति तक पहुँचा जा सकता है|

अतिरिक्त विवरण के लिए देखें [Sokolov *et al.* (q-UCC ansatzes)](https://arxiv.org/abs/1911.10864v2) and [Barkoutsos *et al.* (Heuristic ansatzes)](https://arxiv.org/pdf/1805.04340.pdf)|

### वीक्यूई (VQE)

एक हर्मिटियन ऑपरेटर $\hat H$ को देखते हुए जिसका न्यूनतम eigenvalue $E_{min}$ अज्ञात है और eigenstate $|\psi_{min}\rangle$ से संबद्ध है, VQE एक अनुमान $E_{\theta}$ प्रदान करता है जो $E_{min}$ से सीमित है: 

\begin{align*}
    E_{min} \le E_{\theta} \equiv \langle \psi(\theta) |\hat H|\psi(\theta) \rangle
\end{align*}  

जहां, $|\psi(\theta)\rangle$ $E_{\theta}$ से संबद्ध परीक्षण स्थिति है. $U(\theta)$ द्वारा दर्शाए गए पैरामीटरयुक्त सर्किट को एक आकस्मिक प्रारंभिक अवस्था $|\psi\rangle$ के लिए लागू करके, एल्गोरिथ्म $|\psi_{min}\rangle$ पर एक अनुमान प्राप्त करता है $U(\theta)|\psi\rangle \equiv |\psi(\theta)\rangle$ | शास्त्रीय अनुकूलक $\theta$ पैरामीटर को बदलकर अनुमान को पुनरावृत्त रूप से अनुकूलित करता है और $\langle \psi(\theta) |\hat H|\psi(\theta) \rangle$ की अपेक्षा मूल्य को करता है| 

VQE के अनुप्रयोगों के रूप में, आणविक गतिकी सिमुलेशन में संभावनाएं हैं, मिसाल के तौर पर, देखिए [Sokolov *et al.*, 2021](https://arxiv.org/abs/2008.08144v1), और उत्तेजित स्थितियों की गणना के लिए, देखिये [Ollitrault *et al.*, 2019](https://arxiv.org/abs/1910.12890)|

#### 2. इलेक्ट्रॉनिक संरचना समस्या


फिर आप एक `electronic structure problem` बना सकते हैं जो उन्हें क्वैबिट्स (पॉली स्ट्रिंग्स) में मैप करने से पहले फर्मोनिक ऑपरेटरों की सूची तैयार कर सकता है।

```python
from qiskit_nature.problems.second_quantization.electronic import ElectronicStructureProblem
problem = ElectronicStructureProblem(driver)

# Generate the second-quantized operators
second_q_ops = problem.second_q_ops()

# Hamiltonian
main_op = second_q_ops[0]
```

#### 3. क्यूबिट कन्वर्टर

क्यूबिट कन्वर्टर उस मैपिंग को परिभाषित करने की अनुमति देता है जिसका उपयोग आप सिमुलेशन में करेंगे। आप अलग मैपिंग की कोशिश कर सकते हैं लेकिन
हम एक साधारण पत्राचार की अनुमति के रूप में `Jordan-Wigner` से टिके रहेंगे: एक क्यूबिट अणु में एक स्पिन-कक्षीय का प्रतिनिधित्व करता है।

```python
from qiskit_nature.mappers.second_quantization import ParityMapper, BravyiKitaevMapper, JordanWignerMapper
from qiskit_nature.converters.second_quantization.qubit_converter import QubitConverter

# Setup the mapper and qubit converter
mapper_type = 'JordanWignerMapper'

if mapper_type == 'ParityMapper':
    mapper = ParityMapper()
elif mapper_type == 'JordanWignerMapper':
    mapper = JordanWignerMapper()
elif mapper_type == 'BravyiKitaevMapper':
    mapper = BravyiKitaevMapper()

converter = QubitConverter(mapper=mapper, two_qubit_reduction=False)

# The fermionic operators are mapped to qubit operators
num_particles = (problem.molecule_data_transformed.num_alpha,
             problem.molecule_data_transformed.num_beta)
qubit_op = converter.convert(main_op, num_particles=num_particles)
```

#### 4. प्रारम्भिक अवस्था
जैसा कि हमने थ्योरी सेक्शन में वर्णित किया है, रसायन विज्ञान में एक अच्छी प्रारंभिक अवस्था एचएफ(HF) अवस्था है (अर्थात $|\Psi_{HF} \rangle = |0101 \rangle$). हम इसे इस प्रकार प्रारंभ कर सकते हैं:

```python
from qiskit_nature.circuit.library import HartreeFock

num_particles = (problem.molecule_data_transformed.num_alpha,
             problem.molecule_data_transformed.num_beta)
num_spin_orbitals = 2 * problem.molecule_data_transformed.num_molecular_orbitals
init_state = HartreeFock(num_spin_orbitals, num_particles, converter)
print(init_state)
```

#### 5. अंसात्ज़
आपके द्वारा बनाए जा रहे सबसे महत्वपूर्ण विकल्पों में से एक क्वांटम सर्किट है जिसे आप अपनी जमीनी स्थिति का अनुमान लगाने के लिए चुनते हैं।
यहां किस्किट सर्किट लाइब्रेरी का उदाहरण दिया गया है जिसमें आपका अपना सर्किट बनाने की कई संभावनाएं हैं।

```python
from qiskit.circuit.library import TwoLocal
from qiskit_nature.circuit.library import UCCSD, PUCCD, SUCCD

# Choose the ansatz
ansatz_type = "TwoLocal"

# Parameters for q-UCC antatze
num_particles = (problem.molecule_data_transformed.num_alpha,
             problem.molecule_data_transformed.num_beta)
num_spin_orbitals = 2 * problem.molecule_data_transformed.num_molecular_orbitals

# Put arguments for twolocal
if ansatz_type == "TwoLocal":
    # Single qubit rotations that are placed on all qubits with independent parameters
    rotation_blocks = ['ry', 'rz']
    # Entangling gates
    entanglement_blocks = 'cx'
    # How the qubits are entangled 
    entanglement = 'full'
    # Repetitions of rotation_blocks + entanglement_blocks with independent parameters
    repetitions = 3
    # Skip the final rotation_blocks layer
    skip_final_rotation_layer = True
    ansatz = TwoLocal(qubit_op.num_qubits, rotation_blocks, entanglement_blocks, reps=repetitions, 
                      entanglement=entanglement, skip_final_rotation_layer=skip_final_rotation_layer)
    # Add the initial state
    ansatz.compose(init_state, front=True, inplace=True)
elif ansatz_type == "UCCSD":
    ansatz = UCCSD(converter,num_particles,num_spin_orbitals,initial_state = init_state)
elif ansatz_type == "PUCCD":
    ansatz = PUCCD(converter,num_particles,num_spin_orbitals,initial_state = init_state)
elif ansatz_type == "SUCCD":
    ansatz = SUCCD(converter,num_particles,num_spin_orbitals,initial_state = init_state)
elif ansatz_type == "Custom":
    # Example of how to write your own circuit
    from qiskit.circuit import Parameter, QuantumCircuit, QuantumRegister
    # Define the variational parameter
    theta = Parameter('a')
    n = qubit_op.num_qubits
    # Make an empty quantum circuit
    qc = QuantumCircuit(qubit_op.num_qubits)
    qubit_label = 0
    # Place a Hadamard gate
    qc.h(qubit_label)
    # Place a CNOT ladder
    for i in range(n-1):
        qc.cx(i, i+1)
    # Visual separator
    qc.barrier()
    # rz rotations on all qubits
    qc.rz(theta, range(n))
    ansatz = qc
    ansatz.compose(init_state, front=True, inplace=True)

print(ansatz)
```

#### 8. सटीक ईगेनसॉल्वर

सीखने के उद्देश्यों के लिए, हम हैमिल्टनियन मैट्रिक्स के सटीक विकर्ण के साथ समस्या को ठीक से हल कर सकते हैं ताकि हम जान सकें कि वीक्यूई के द्वारा कहां लक्ष्य करना है। 
बेशक, इस मैट्रिक्स के आयाम आणविक कक्षाओं की संख्या में तेजी से बढ़ते हैं ताकि आप अपनी पसंद के बड़े अणु के लिए ऐसा करने का प्रयास कर सकें और देखें कि यह कितना धीमा हो जाता है।
